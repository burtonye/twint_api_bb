```python
import pandas as pd

# df = pd.read_csv('smcounty_twitter.csv')
df = pd.read_csv('twitterallppl-8071174939.csv')
df['handle'] = df['External ID'].str[23:]
targets_list_cleaned = df['handle'].to_list()
targets_list_cleaned1 = targets_list_cleaned[1800:]
```

```python
df2 = pd.DataFrame()

for idx, target in enumerate(targets_list_cleaned1):
        try:
            user = api.get_user(target)
            df_temp = pd.DataFrame(
                [[target, user.name, user.followers_count, user.verified]]
            )
            df2=df2.append(df_temp)
            print (idx, target, user.name, user.followers_count, user.verified)
        except tweepy.TweepError as e:
            df_temp = pd.DataFrame([[target,'error',0,'false']])
            df2=df2.append(df_temp)
            print(idx, target, 'error')
            pass
```

```python
df2.columns = ['handle','name','followers','verified']
df.set_index('handle').join(df2.set_index('handle')).to_excel("all_twitter.xlsx")
```
